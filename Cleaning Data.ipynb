```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
```

```python
dataset = pd.read_csv('Adidas US Sales Datasets.csv')
dataset
```

```python
def year(x):
  return x[-4:]
dataset['Year'] = dataset['Invoice Date'].apply(year)
```

```python
dataset.columns
```

```python
def cleaning(x):
  y=''
  for i in x:
    if i.isdigit()==True:
      y += i
    elif i == '.':
      y += i
  return float(y)

dataset['Price per Unit'] = dataset['Price per Unit'].apply(cleaning)
dataset['Total Sales'] = dataset['Total Sales'].apply(cleaning)
dataset['Units Sold'] = dataset['Units Sold'].apply(cleaning)
dataset['Operating Profit'] = dataset['Operating Profit'].apply(cleaning)
```

```python
dataset[['Price per Unit', 'Total Sales', 'Units Sold', 'Operating Profit']].head(5)
```

```python
print('---Units Sold---')
max = dataset['Units Sold'].max()
min = dataset['Units Sold'].min()
print(f'Maximum : {max}')
print(f'Minimum : {min}\n')

print('---Operating Profit---')
max = dataset['Operating Profit'].max()
min = dataset['Operating Profit'].min()
print(f'Maximum : {max}')
print(f'Minimum : {min}\n')

print('---Total Sales---')
max = dataset['Total Sales'].max()
min = dataset['Total Sales'].min()
print(f'Maximum : {max}')
print(f'Minimum : {min}\n')

print('---Price per Unit---')
max = dataset['Price per Unit'].max()
min = dataset['Price per Unit'].min()
print(f'Maximum : {max}')
print(f'Minimum : {min}\n')

print('---Operating Margin---')
max = dataset['Operating Margin'].max()
min = dataset['Operating Margin'].min()
print(f'Maximum : {max}')
print(f'Minimum : {min}\n')
```

```python
Total = dataset['Units Sold'].sum()
print(f'Total Units Sold : {Total}')
Total = dataset['Operating Profit'].sum()
print(f'Total Operating Profit : {Total}')
Total = dataset['Total Sales'].sum()
print(f'Total Sales : {Total}')
```

```python
def gender(x):
    if 'Men' in x:
        return 'Men'
    elif 'Women' in x:
        return 'Women'

dataset['Gender'] = dataset['Product'].apply(gender)
```

```python
def type(x):
    if 'Footwear' in x:
        return 'Footwear'
    elif 'Apparel' in x:
        return 'Apparel'

dataset['Type'] = dataset['Product'].apply(type)
```

```python
item = dataset[['Region', 'Total Sales']]
item = item.groupby('Region')[['Total Sales']].sum().reset_index()

lowest_region = item.loc[item['Total Sales'].idxmin()]
highest_region = item.loc[item['Total Sales'].idxmax()]

print(f"The region with the lowest total sales is {lowest_region['Region']} with {lowest_region['Total Sales']} total sales.")
print(f"The region with the highest total sales is {highest_region['Region']} with {highest_region['Total Sales']} total sales.")

plt.bar(item['Region'], item['Total Sales'])
plt.xlabel("Region")
plt.ylabel('Total Sales')
plt.show()
```

```python
item = dataset[['Region','Operating Profit']]
item = item.groupby('Region').agg({'Operating Profit': 'sum'}).reset_index()

lowest_region_profit = item.loc[item['Operating Profit'].idxmin()]
highest_region_profit = item.loc[item['Operating Profit'].idxmax()]

print(f"The region with the lowest operating profit is {lowest_region_profit['Region']} with {lowest_region_profit['Operating Profit']} operating profit.")
print(f"The region with the highest operating profit is {highest_region_profit['Region']} with {highest_region_profit['Operating Profit']} operating profit.")

plt.bar(item['Region'], item['Operating Profit'])
plt.xlabel("Region")
plt.ylabel('Operating Profit')
plt.show()
```

```python
dataset['Invoice Date'] = pd.to_datetime(dataset['Invoice Date'], format="%d-%m-%Y")
dataset.set_index('Invoice Date', inplace=True)

quarterly_sales = dataset['Total Sales'].resample('Q').sum()
quarterly_profit = dataset['Operating Profit'].resample('Q').sum()

plt.figure(figsize=(12, 6))
plt.plot(quarterly_sales, label='Total Sales')
plt.plot(quarterly_profit, label='Operating Profit')


for i, sales in enumerate(quarterly_sales):
    plt.annotate(f'{sales:.2f}', (quarterly_sales.index[i], sales), textcoords="offset points", xytext=(0, 5), ha='center')

for i, profit in enumerate(quarterly_profit):
    plt.annotate(f'{profit:.2f}', (quarterly_profit.index[i], profit), textcoords="offset points", xytext=(0, -10), ha='center', color='blue')

plt.title('Quarterly Sales and Operating Profit Trends')
plt.xlabel('Quarter')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

```python
product_performance = dataset.groupby('Product')['Total Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
product_performance.plot(kind='bar', color='skyblue')
plt.title('Product Performance')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
regional_performance = dataset.groupby('Region')['Total Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
regional_performance.plot(kind='bar', color='salmon')
plt.xticks(rotation=75)
plt.title('Regional Performance')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
state_city_profitability = dataset.groupby(['State', 'City'])['Operating Profit'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
state_city_profitability.nlargest(10).plot(kind='bar', color='green')
plt.title('Top 10 States/Cities by Operating Profit')
plt.xlabel('State/City')
plt.ylabel('Operating Profit')
plt.show()
```
